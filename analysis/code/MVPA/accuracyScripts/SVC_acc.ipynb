```python
def SS_heatmaps(subid):
    import pandas as pd
    import numpy as np
    import seaborn as sns
    import matplotlib as plt
    %matplotlib inline
    df=pd.read_csv('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/results/SVC/acc/SS/' +subid+ '_accTable_same_sub.csv')
    df.set_index('Test_variables', inplace=True)
    ax=sns.heatmap(df, annot=True)
    fig=ax.get_figure()
    fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/SVC/acc/SS/' +subid+ '_acc.png', bbox_inches='tight')
    
```

```python
def SS_concat():
    import pandas as pd
    import numpy as np
    import seaborn as sns
    import matplotlib as plt
    %matplotlib inline
    #concatenate all dataframes
    accTable=[]
    subList=['MSC01','MSC02','MSC03','MSC04','MSC05','MSC06','MSC07','MSC10']
    taskList=['mem','mixed','motor']
    for i in subList:
        df=pd.read_csv('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/results/SVC/acc/SS/' +i+ '_accTable_same_sub.csv')
        df['sub']=i
        accTable.append(df)
    results=pd.concat(accTable)
    grouped_df=results.groupby('Test_variables')
    for j in taskList:
        task_df=grouped_df.get_group(j)
        task_df.set_index('sub', inplace=True)
        task_df.drop(columns=[j, 'Test_variables'], inplace=True)
        ax=sns.heatmap(task_df, annot=True)
        ax.set_title('Testing ' + j)
        fig=ax.get_figure()
        fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/SVC/acc/SS/' +j+ '_acc_heatmap.png', bbox_inches='tight')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
#concatenate all dataframes
accTable=[]
subList=['MSC01','MSC02','MSC03','MSC04','MSC05','MSC06','MSC07','MSC10']
for i in subList:
    df=pd.read_csv('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/results/SVC/acc/SS/'+i+'_accTable_same_sub.csv')
    df['sub']=i
    accTable.append(df)
results=pd.concat(accTable)
#groupby the testing variables
grouped_df=results.groupby('Test_variables')
memTest_df=grouped_df.get_group('mem')
memTest_df.set_index('sub', inplace=True)
memTest_df.drop(columns=['mem', 'Test_variables'], inplace=True)


mixedTest_df=grouped_df.get_group('mixed')
mixedTest_df.set_index('sub', inplace=True)
mixedTest_df.drop(columns=['mixed','Test_variables'], inplace=True)


motorTest_df=grouped_df.get_group('motor')
motorTest_df.set_index('sub', inplace=True)
motorTest_df.drop(columns=['motor','Test_variables'], inplace=True)
```

```python
ax=sns.heatmap(memTest_df, annot=True)

ax.set_title('Testing Memory')
fig=ax.get_figure()
fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/SVC/acc/SS/mem_acc_heatmap.png', bbox_inches='tight')
```

```python
ax=sns.heatmap(mixedTest_df, annot=True)

ax.set_title('Testing Mixed')
fig=ax.get_figure()
fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/SVC/acc/SS/mixed_acc_heatmap.png', bbox_inches='tight')
```
