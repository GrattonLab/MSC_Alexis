```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
#concatenate all dataframes
accTable=[]
subList=['MSC01','MSC02','MSC03','MSC04','MSC05','MSC06','MSC07','MSC08','MSC10']
for i in subList:
    df=pd.read_csv('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/results/logReg/acc/SS/'+i+'_accTable_same_sub.csv')
    df['sub']=i
    accTable.append(df)
results=pd.concat(accTable)
#groupby the testing variables
grouped_df=results.groupby('Test_variables')
memTest_df=grouped_df.get_group('mem')
memTest_df.set_index('sub', inplace=True)
memTest_df.drop(columns=['mem', 'Test_variables'], inplace=True)


mixedTest_df=grouped_df.get_group('mixed')
mixedTest_df.set_index('sub', inplace=True)
mixedTest_df.drop(columns=['mixed','Test_variables'], inplace=True)


motorTest_df=grouped_df.get_group('motor')
motorTest_df.set_index('sub', inplace=True)
motorTest_df.drop(columns=['motor','Test_variables'], inplace=True)
```

```python
ax=sns.heatmap(memTest_df, annot=True)

ax.set_title('Testing Memory')
fig=ax.get_figure()
fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/logReg/acc/SS/mem_acc_heatmap.png', bbox_inches='tight')
```

```python
ax=sns.heatmap(mixedTest_df, annot=True)

ax.set_title('Testing Mixed')
fig=ax.get_figure()
fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/logReg/acc/SS/mixed_acc_heatmap.png', bbox_inches='tight')
```

```python
ax=sns.heatmap(motorTest_df, annot=True)

ax.set_title('Testing Motor')
fig=ax.get_figure()
fig.savefig('/Users/aporter1350/Desktop/MSC_Alexis/analysis/output/images/logReg/acc/SS/motor_acc_heatmap.png', bbox_inches='tight')
```
