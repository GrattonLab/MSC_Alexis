```python
#Code to help get all csv files into a useable format for making into figures
import pandas as pd
import numpy as np
import sys
import os
thisDir = os.path.expanduser('~/Desktop/MSC_Alexis/analysis/')
#directory for all regular analysis
outDir = thisDir + 'output/results/'
taskList=['semantic','mem','motor','glass']
subList=['MSC01','MSC02','MSC03','MSC04','MSC05','MSC06','MSC07','MSC10']
#all permutations of features
subDir=outDir+'subNetwork/'
randDir=outDir+'rdmNetwork/'
netList=['unassign','default', 'visual','fp','dan','van','salience','co','sm','sm-lat', 'auditory','pmn','pon']
feature=[14808,10824,8736, 4620,5264,3151, 484, 4060, 2375,316, 564,45,21]
netRoi=dict([('unassign',14808),('default', 10824),('visual',8736),('fp', 4620),('dan',5264),('van',3151),('salience', 484),('co', 4060),('sm', 2375),('sm-lat', 316),('auditory', 564),('pmn',45),('pon',21)])

#will change to new feature selection when done on quest

rdmDir = '~/Desktop/MSC_Alexis/analysis/output/results/rdmNetwork/'
rand_DS=pd.read_csv(rdmDir+'DS/acc.csv')
rand_SS=pd.read_csv(rdmDir+'SS/acc.csv')
rand_BS=pd.read_csv(rdmDir+'BS/acc.csv')
rand_CV=pd.read_csv(rdmDir+'CV/acc.csv')

lower_DS=pd.read_csv(rdmDir+'DS/Min_acc.csv')
lower_SS=pd.read_csv(rdmDir+'SS/Min_acc.csv')
lower_BS=pd.read_csv(rdmDir+'BS/Min_acc.csv')
lower_CV=pd.read_csv(rdmDir+'CV/Min_acc.csv')

rand_DS=pd.concat([rand_DS, lower_DS])
rand_SS=pd.concat([rand_SS, lower_SS])
rand_BS=pd.concat([rand_BS, lower_BS])
rand_CV=pd.concat([rand_CV, lower_CV])

#Random sampling and feature selection calculate z scores
#subnetworks overlapped with random network sampling
SS_nets=pd.DataFrame()
DS_nets=pd.DataFrame()
CV_nets=pd.DataFrame()
BS_nets=pd.DataFrame()
for i in netList:
    tmpSS=pd.read_csv(subDir+i+'/SS/acc.csv')
    tmpSS['Network']=i
    tmpSS['Features']=netRoi[i]
    SS_nets=pd.concat([SS_nets,tmpSS])
    
    tmpDS=pd.read_csv(subDir+i+'/DS/acc.csv')
    tmpDS['Network']=i
    tmpDS['Features']=netRoi[i]
    DS_nets=pd.concat([DS_nets,tmpDS])
    
    tmpBS=pd.read_csv(subDir+i+'/BS/acc.csv')
    tmpBS['Network']=i
    tmpBS['Features']=netRoi[i]
    BS_nets=pd.concat([BS_nets,tmpBS])
    
    tmpCV=pd.read_csv(subDir+i+'/CV/acc.csv')
    tmpCV=tmpCV.melt(id_vars='sub', value_vars=['glass','semantic', 'motor','mem'], var_name='Task', value_name='Accuracy')
    tmpCV['Network']=i
    tmpCV['Features']=netRoi[i]
    CV_nets=pd.concat([CV_nets,tmpCV])
#extra co/co co/fp and cp/dmn analysis    
blockList=['co_co','default_co','fp_co']
interBlock=dict([('co_co',780),('fp_co',960),('default_co',1640)])
for block in blockList:    
    tmpSS=pd.read_csv(subDir+block+'/SS/acc.csv')
    tmpSS['Network']=block
    tmpSS['Features']=interBlock[block]
    SS_nets=pd.concat([SS_nets,tmpSS])
    
    tmpDS=pd.read_csv(subDir+block+'/DS/acc.csv')
    tmpDS['Network']=block
    tmpDS['Features']=interBlock[block]
    DS_nets=pd.concat([DS_nets,tmpDS])
    
    tmpBS=pd.read_csv(subDir+i+'/BS/acc.csv')
    tmpBS['Network']=i
    tmpBS['Features']=interBlock[block]
    BS_nets=pd.concat([BS_nets,tmpBS])
    
    tmpCV=pd.read_csv(subDir+block+'/CV/acc.csv')
    tmpCV.rename(columns={'acc':'Accuracy','task':'Task'},inplace=True)
    tmpCV.drop(columns='Unnamed: 0', inplace=True)
    #tmpCV=tmpCV.melt(id_vars='sub', value_vars=['glass','semantic', 'motor','mem'], var_name='Task', value_name='Accuracy')
    tmpCV['Network']=block
    tmpCV['Features']=interBlock[block]
    CV_nets=pd.concat([CV_nets,tmpCV])   
    
SS_nets.drop(columns='Unnamed: 0', inplace=True)
DS_nets.drop(columns='Unnamed: 0', inplace=True)
BS_nets.drop(columns='Unnamed: 0', inplace=True)
CV_nets.rename(columns={'Accuracy':'acc'},inplace=True)

#additional fp/co/dmn relationship had to update list
newNetList=['unassign','default', 'visual','fp','dan','van','salience','co','sm','sm-lat', 'auditory','pmn','pon','co_co','default_co','fp_co']
newFeat=[14808,10824,8736, 4620,5264,3151, 484, 4060, 2375,316, 564,45,21,780,1640,960]
#netRoi
SStotal=[]
DStotal=[]
CVtotal=[]
BStotal=[]
for i in newNetList:
    #CV
    CVtmp=CV_nets[(CV_nets['Network']==i)].acc.values
    CVacc=CVtmp.mean()
    CVtotal.append(CVacc)
    #SS
    SStmp=SS_nets[(SS_nets['Network']==i)].acc.values
    SSacc=SStmp.mean()
    SStotal.append(SSacc)
    #DS
    DStmp=DS_nets[(DS_nets['Network']==i)].acc.values
    DSacc=DStmp.mean()
    DStotal.append(DSacc)
    #BS
    BStmp=BS_nets[(BS_nets['Network']==i)].acc.values
    BSacc=BStmp.mean()
    BStotal.append(BSacc)
    
CV_mu=pd.DataFrame({'Network':newNetList, 'Acc':CVtotal,'features':newFeat})
SS_mu=pd.DataFrame({'Network':newNetList, 'Acc':SStotal,'features':newFeat})
DS_mu=pd.DataFrame({'Network':newNetList, 'Acc':DStotal,'features':newFeat})
BS_mu=pd.DataFrame({'Network':newNetList, 'Acc':BStotal,'features':newFeat})

```

```python
#same sub same task
tmp_mu=[]
tmp_std=[]
for f in feature:
    subset_rand_CV=rand_CV[rand_CV['features']==f]
    z_mu_tmp=subset_rand_CV['acc'].mean()
    z_std_tmp=subset_rand_CV['acc'].std()
    tmp_mu.append(z_mu_tmp)
    tmp_std.append(z_std_tmp)
cv_z=pd.DataFrame({'mu':tmp_mu, 'std':tmp_std, 'features':feature})
CV_zscores=CV_mu.merge(cv_z,how='left',on=['features'])
CV_zscores['z_score']=(CV_zscores['Acc']-CV_zscores['mu'])/CV_zscores['std']

#same sub diff task
tmp_mu=[]
tmp_std=[]
for f in feature:
    subset_rand_SS=rand_SS[rand_SS['features']==f]
    z_mu_tmp=subset_rand_SS['acc'].mean()
    z_std_tmp=subset_rand_SS['acc'].std()
    tmp_mu.append(z_mu_tmp)
    tmp_std.append(z_std_tmp)
SS_z=pd.DataFrame({'mu':tmp_mu, 'std':tmp_std, 'features':feature})
SS_zscores=SS_mu.merge(SS_z,how='left',on=['features'])
SS_zscores['z_score']=(SS_zscores['Acc']-SS_zscores['mu'])/SS_zscores['std']

#diff sub same task
tmp_mu=[]
tmp_std=[]
for f in feature:
    subset_rand_DS=rand_DS[rand_DS['features']==f]
    z_mu_tmp=subset_rand_DS['acc'].mean()
    z_std_tmp=subset_rand_DS['acc'].std()
    tmp_mu.append(z_mu_tmp)
    tmp_std.append(z_std_tmp)
DS_z=pd.DataFrame({'mu':tmp_mu, 'std':tmp_std, 'features':feature})
DS_zscores=DS_mu.merge(DS_z,how='left',on=['features'])
DS_zscores['z_score']=(DS_zscores['Acc']-DS_zscores['mu'])/DS_zscores['std']

#diff sub diff task
tmp_mu=[]
tmp_std=[]
for f in feature:
    subset_rand_BS=rand_BS[rand_BS['features']==f]
    z_mu_tmp=subset_rand_BS['acc'].mean()
    z_std_tmp=subset_rand_BS['acc'].std()
    tmp_mu.append(z_mu_tmp)
    tmp_std.append(z_std_tmp)
BS_z=pd.DataFrame({'mu':tmp_mu, 'std':tmp_std, 'features':feature})
BS_zscores=BS_mu.merge(BS_z,how='left',on=['features'])
BS_zscores['z_score']=(BS_zscores['Acc']-BS_zscores['mu'])/BS_zscores['std']




```

```python
CV_zscores.to_csv(randDir+'CV/zscore_all.csv',index=False)
SS_zscores.to_csv(randDir+'SS/zscore_all.csv',index=False)
DS_zscores.to_csv(randDir+'DS/zscore_all.csv',index=False)
BS_zscores.to_csv(randDir+'BS/zscore_all.csv',index=False)


rand_DS.to_csv(rdmDir+'DS/new_acc.csv',index=False)
rand_SS.to_csv(rdmDir+'SS/new_acc.csv',index=False)
rand_BS.to_csv(rdmDir+'BS/new_acc.csv',index=False)
rand_CV.to_csv(rdmDir+'CV/new_acc.csv',index=False)

```
