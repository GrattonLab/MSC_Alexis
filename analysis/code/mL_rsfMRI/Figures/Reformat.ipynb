```python
SS=NetSS.groupby(['feature','network']).mean()
SS_z=rdmNetSS.groupby(['features','Null_Network']).mean()
SS_z['std']=rdmNetSS.groupby(['features','Null_Network']).std()
SS_z.reset_index(inplace=True)
SS_z.rename(columns={'features':'feature','Null_Network':'network','acc':'mu'},inplace=True)
SS_zscores=SS.merge(SS_z,how='left',on=['network'])
SS_zscores['z_score']=(SS_zscores['acc']-SS_zscores['mu'])/SS_zscores['std']
```

```python
BS=NetBS.groupby(['feature','network']).mean()
BS_z=rdmNetBS.groupby(['features','Null_Network']).mean()
BS_z['std']=rdmNetBS.groupby(['features','Null_Network']).std()
BS_z.reset_index(inplace=True)
BS_z.rename(columns={'features':'feature','Null_Network':'network','acc':'mu'},inplace=True)
BS_zscores=BS.merge(BS_z,how='left',on=['network'])
BS_zscores['z_score']=(BS_zscores['acc']-BS_zscores['mu'])/BS_zscores['std']
```
