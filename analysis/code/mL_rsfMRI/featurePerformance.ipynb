```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os
CV=pd.DataFrame()
SS=pd.DataFrame()
DS=pd.DataFrame()
BS=pd.DataFrame()
thisDir = os.path.expanduser('~/Desktop/MSC_Alexis/analysis/')
outDir= thisDir + 'output/randomFeat/'
features=[100,500,1000,1500,2000,2500,3000,4000,5000,10000,15000,20000,30000,40000,50000]
for feat in features:
    #CV
    CVfile=pd.read_csv(outDir+'/results/ridge/acc/CV/'+str(feat)+'.csv')
    CVfile.drop(columns='sub', inplace=True)
    CVfile=CVfile.melt(value_vars=['mixed', 'motor','mem'], var_name='Task', value_name='Accuracy')
    CVfile['Features']=feat
    CV=pd.concat([CV,CVfile])
    #SS
    SSfile=pd.read_csv(outDir+'/results/ridge/acc/SS/'+str(feat)+'.csv')
    SSfile['Features']=feat
    SS=pd.concat([SS,SSfile])
    #DS
    DSfile=pd.read_csv(outDir+'/results/ridge/acc/DS/'+str(feat)+'.csv')
    DSfile['Features']=feat
    DS=pd.concat([DS,DSfile])
    #BS
    BSfile=pd.read_csv(outDir+'/results/ridge/acc/BS/'+str(feat)+'.csv')
    BSfile['Features']=feat
    BS=pd.concat([BS,BSfile])
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Features", y="acc",data=SS)
ax.set_title('Same Subject Different Task')
ax.set_ylabel('Performance(%)')
ax.axhline(.50, ls='--', color='r')
ax.set(ylim=(0, 1))
plt.savefig(outDir +'/results/ridge/acc/SS/line_SS.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Features", y="acc",data=DS)
ax.set_title('Different Subject Same Task')
ax.set_ylabel('Performance(%)')
ax.axhline(.50, ls='--', color='r')
ax.set(ylim=(0, 1))
plt.savefig(outDir +'/results/ridge/acc/DS/line_DS.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Features", y="acc",data=BS)
ax.set_title('Different Subject Different Task')
ax.set_ylabel('Performance(%)')
ax.axhline(.50, ls='--', color='r')
ax.set(ylim=(0, 1))
plt.savefig(outDir +'/results/ridge/acc/BS/line_BS.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Features", y="Accuracy",data=CV)
ax.set_title('Cross Validation')
ax.set_ylabel('Performance(%)')
ax.axhline(.50, ls='--', color='r')
ax.set(ylim=(0, 1))
plt.savefig(outDir +'/results/ridge/acc/CV/line_CV.png', bbox_inches='tight')
```
